# **Thực hiện các yêu cấu sau :**
1. Vẽ đồ thị scatter plot thể hiện mối tương quan giữa postmolt và premolt
2. Tính hệ số tương quan giữa postmolt và premolt
3. Xây dựng hồi quy tuyến tính
4. Kiểm định phương trình hồi quy tuyến tính 
5. Dựa vào phương trình hồi quy đã xây dựng để dự đoán

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from scipy.stats.stats import pearsonr
import statsmodels.formula.api as smf
from scipy import stats
import seaborn as sns
%matplotlib inline
```

```python
df_hq = pd.read_csv('https://raw.githubusercontent.com/huynhhoc/DataAnalystDeepLearning/main/Data/crabs.csv',sep = '\s+')
df_hq.head()
```

# **1. Vẽ đồ thị scatter plot thể hiện mối tương quan giữa postmolt và premolt**

```python
plt.plot(df_hq['presz'],df_hq['postsz'], 'o',linewidth=2);
plt.xlabel('Postmost size')
plt.ylabel('Premolt size')
plt.title('Postmost vs Premolt')
```

# **2. Tính hệ số tương quan giữa postmolt và premolt**

```python
pearsonr(df_hq['postsz'],df_hq['presz'])
```

# **3. Xây dựng phương trình hồi quy tuyến tính**

```python
result = smf.ols('presz~postsz', df_hq).fit()
print('Du Bao', result.summary())
```

# **4. Kiểm định phương trình hồi quy tuyến tính**

Dựa trên phương trình hồi tuyến tính : 
premost= -25.2137+1.0732*postmolt

# **5.Dùng phương trình hồi quy để dự báo**

```python
sales_pred = result.predict() # dự đoán
plt.plot(df_hq['postsz'],df_hq['presz'],'o', color = 'b') 
plt.plot(df_hq['postsz'],sales_pred,'r',linewidth=2) 
plt.xlabel('Postmolt size')
plt.ylabel('Premolt size')
plt.title('Premolt and Postmolt ')
plt.show()
```
